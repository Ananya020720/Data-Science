```python
import pandas as pd
import matplotlib.pyplot as plt

sales_data = pd.read_csv('sales_data.csv')

# Group by region and sum total revenue
revenue_by_region = sales_data.groupby('Region')['Total Revenue'].sum()

# Plot as pie chart
plt.figure(figsize=(8, 8))
plt.pie(revenue_by_region, labels=revenue_by_region.index, autopct='%1.1f%%', startangle=140, colors=['blue','red','green'])
plt.title('Revenue Contribution by Region')
plt.tight_layout()
plt.show()
```
