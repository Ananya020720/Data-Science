```python
# Identify the top 5 products with the highest total revenue.

import matplotlib.pyplot as plt
import pandas as pd

# Load and process data
sales_data = pd.read_csv('sales_data.csv')

# Aggregate total revenue per product
top_products = sales_data.groupby('Product Name')['Total Revenue'].sum().reset_index()

# Sort in descending order and grab the top 5
top_5 = top_products.sort_values(by='Total Revenue', ascending=False).head(5)
print(top_5)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(top_5['Product Name'], top_5['Total Revenue'], color='skyblue')
plt.title('Top 5 Selling Products by Total Revenue')
plt.xlabel('Product Name')
plt.ylabel('Total Revenue')
plt.tight_layout()
plt.show()
```
