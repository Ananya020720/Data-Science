```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
sales_data = pd.read_csv('sales_data.csv')

# Group by category and calculate the mean of unit price
avg_price_category = sales_data.groupby('Product Category')['Unit Price'].mean().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(avg_price_category['Product Category'], avg_price_category['Unit Price'], color='black')
plt.title('Average Unit Price per Product Category')
plt.xlabel('Average Unit Price')
plt.ylabel('Product Category')
plt.tight_layout()
plt.show()
```
