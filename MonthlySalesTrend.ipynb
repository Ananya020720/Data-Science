```python
# Convert the Date column to datetime and plot monthly total revenue

import matplotlib.pyplot as plt
import pandas as pd

# Read and process the CSV file
sales_data = pd.read_csv('sales_data.csv')
sales_data['Date'] = pd.to_datetime(sales_data['Date'])

# Extract year and month for grouping
sales_data['YearMonth'] = sales_data['Date'].dt.to_period('M')

# Group by Year-Month and sum revenue
monthly_revenue = sales_data.groupby('YearMonth')['Total Revenue'].sum().reset_index()

# Convert YearMonth back to datetime for plotting
monthly_revenue['YearMonth'] = monthly_revenue['YearMonth'].dt.to_timestamp()

# Plot the trend
plt.figure(figsize=(10, 5))
plt.plot(monthly_revenue['YearMonth'], monthly_revenue['Total Revenue'], marker='o')
plt.title('Monthly Total Revenue')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.grid(True)
plt.show()
```
